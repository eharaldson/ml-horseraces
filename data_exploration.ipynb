```python
import pandas as pd
import numpy as np 
```

```python
df = pd.read_csv('ProcessedDatasetNulls.csv', index_col=0)
```

```python
df = df.drop(columns=['win_number_by_HorseDistance', 'avg_place_last_3_by_HorseGoing',
       'win_last_5_by_HorseJockey', 'num_races_by_HorseDistance',
       'avg_place_by_HorseJockey', 'win_last_3_by_HorseJockey',
       'win_rate_by_HorseGoing', 'win_rate_by_HorseJockey',
       'win_rate_by_HorseDistance', 'OffR',
       'placeTop3_number_last_5_by_HorseGoing',
       'avg_place_last_3_by_HorseDistance', 'avg_place_last_5_by_HorseJockey',
       'win_last_3_by_HorseDistance', 'win_number_by_HorseJockey',
       'win_number_by_HorseGoing', 'placeTop3_number_last_5_by_HorseJockey',
       'num_races_by_HorseJockey', 'win_last_5_by_HorseDistance',
       'avg_place_last_5_by_HorseDistance', 'placetop3_rate_by_HorseJockey',
       'avg_place_last_5_by_HorseGoing', 'win_last_3_by_HorseGoing',
       'placetop3_rate_by_HorseGoing', 'avg_place_by_HorseDistance',
       'placeTop3_number_last_5_by_HorseDistance',
       'avg_place_last_3_by_HorseJockey', 'win_last_5_by_HorseGoing',
       'avg_place_by_HorseGoing', 'num_races_by_HorseGoing',
       'placetop3_rate_by_HorseDistance'])
```

```python
# Clean up null values
for column in df.columns:
    if 'win_last' in column or 'win_num' in column or 'num_races' in column or 'PlaceTop3_number':
        df[column].fillna(0, inplace=True)

    elif 'TimeSince' in column or 'OffR' in column or 'Pace' in column:
        df[column].fillna(df[column].median(), inplace=True)

    else:
        df[column].fillna(df[column].mean(), inplace=True)
```

```python
df
```

```python
for column in df.columns:
    if df[column].dtype != np.float64 and column not in {'Place', 'PlaceInt', 'win', 'Runners', 'DateTime', 'Track Name'}:
        df[column] = df[column].apply(lambda x: True if x == 'True' else False)
```

```python
df.to_csv('ProcessedDatasetExtra.csv')
```
